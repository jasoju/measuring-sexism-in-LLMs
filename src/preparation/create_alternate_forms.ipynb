```python
# load df with OG items
ASI_df = pd.read_csv("../data/ASI.csv", sep=";", index_col=0)
# extract list of items
ASI_items = ASI_df.item.to_list()
```

```python
# loop through all items and generate an alternate form using LLama 3
afs = []
for item in ASI_items:
    messages = [{"role": "user", "content": f"Return an alternate form for the following item of the Ambivalent Sexism Inventory: '{item}'. The alternate form should be different but have an equivalent meaning. Only return the alternate form. Alternate form:"}]
    client = InferenceClient("meta-llama/Meta-Llama-3-8B-Instruct")
    output = client.chat_completion(messages, max_tokens=100)
    af = output.choices[0].message.content
    af = af.replace('"', '')
    afs.append(af)
print(afs)
```

```python
ASI_df["alternate form"] = afs
```
